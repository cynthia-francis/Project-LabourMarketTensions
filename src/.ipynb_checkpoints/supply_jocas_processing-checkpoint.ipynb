````python
#¬†IMPORT LIBRARIES
import matplotlib.pyplot as plt
import pandas as pd
import os
import re #¬†For regular expression
import geopandas as gpd # To read geospatial data
from pathlib import Path # To set relative paths
import unidecode # To standardize strings
import py7zr # To unzip files
import datetime
import csv
````

````python
# GETTING PROJECT'S ROOT DIRECTORY
base_folder = Path().resolve()  # CURRENT WORKING DIRECTORY
main_folder = base_folder.parent

# DIRECTORIES
jocas_dir = "/Users/alfonso/Desktop/JOCAS"
output_path =  main_folder / "data" / "2- Formatted Data" / "jocas_offers_processed.csv"

#¬†IMPORTING FILES
commune_names_path = main_folder / "data" / "2- Formatted Data" / "name_communes_5000.csv"
commune_names = pd.read_csv(commune_names_path).squeeze().tolist()  
rome_fap_path = main_folder / "data" / "linking tables" / "Rome_to_Fap_processed.csv"
rome_fap = pd.read_csv(rome_fap_path)
````

## üèóÔ∏è 1- DATA PROCESSING: AGGREGATING JOB OFFERS

This section automates the extraction, transformation, and aggregation of job offers data from multiple sources. The processing pipeline follows these key steps:

#### **Step 1: Loop through source folders and monthly subfolders**
- The script scans multiple **data source folders** (e.g., `apec`, `regionsjob_nord`) containing job listings.
- Within each source folder, it processes **monthly subfolders** formatted as `YYYY-MM` (e.g., `2024-01`).

#### **Step 2: Read and clean raw job offers data**
- Each `.csv` file inside the monthly subfolders is **read and processed**.
- Key transformations include:
  - **Renaming columns** (`location_label ‚Üí commune`, `job_ROME_code ‚Üí rome_code`).
  - **Standardizing commune names** (lowercase, remove accents, unify "Saint").
  - **Filtering for communes with >5,000 inhabitants** (to keep relevant locations).
  - **Removing missing ROME job codes** (to ensure valid job categories).

#### **Step 3: Merge with job category classifications**
- Each job offer's **ROME code** is mapped to its **FAP87 category** using a linking table.

#### **Step 4: Aggregate job offers by commune, job category, and month**
- The dataset is **grouped by `commune`, `fap87`, and `month`**, summing job offers.

#### **Step 5: Save the final processed dataset**
- All processed files are combined into a **single final dataset**.
- The output is saved as an Excel file: `final_job_offers.xlsx`.

This structured approach ensures that job offers data from different sources and time periods are harmonized into a **single, aggregated dataset** for further analysis. üöÄ


````python
%%capture

# Create a list to store the results
jocas_list = []

# Walk through each data source folder (e.g., apec, regionsjob_nord)
for source_folder in os.listdir(jocas_dir):
    source_path = os.path.join(jocas_dir, source_folder)  # Use the correct root folder variable
    if os.path.isdir(source_path):  # Check if it is a folder

        # Loop through each monthly subfolder within the data source folder
        for month_folder in os.listdir(source_path):
            month_path = os.path.join(source_path, month_folder)

            if os.path.isdir(month_path):  # Check if it is a folder

                for file_name in os.listdir(month_path):
                    if file_name.endswith('.csv'):  # Process only CSV files
                        file_path = os.path.join(month_path, file_name)

                        # Read CSV with robust error handling for inconsistent rows and mixed types
                        df = pd.read_csv(
                            file_path,
                            header=0,
                            sep=';',  # Specify semicolon as the delimiter
                            on_bad_lines='skip',  # Skip problematic rows
                            low_memory=False,     # Avoid dtype warnings
                            quoting=csv.QUOTE_MINIMAL,
                            escapechar='\\'
                        )

                        # RENAME COLUMNS OF INTEREST
                        df.rename(columns={"location_label": "commune", "job_ROME_code": "rome_code"}, inplace=True)

                        # COMMUNE NAME STANDARDIZATION FUNCTION
                        def standardize_commune(name):
                            if pd.isna(name):
                                return None
                            name = unidecode.unidecode(name.lower().strip())  # Remove accents & lowercase
                            name = re.sub(r"[-'‚Äô]", " ", name)  # Remove hyphens & apostrophes
                            name = re.sub(r"\bst[ .]", "saint ", name)  # Standardize "St." -> "Saint"
                            return name

                        # APPLY STANDARDIZATION
                        df["commune"] = df["commune"].apply(standardize_commune)

                        # FILTER FOR COMMUNES WITH MORE THAN 5000 HABITANTS
                        df = df[df["commune"].isin(commune_names)]

                        # DROP OBSERVATIONS WITHOUT ROME_CODE
                        df.dropna(subset=["rome_code"], inplace=True)

                        # MERGE ROME CODE WITH LINKING TABLE (TO FAP)
                        df = df.merge(rome_fap, left_on="rome_code", right_on="rome", how="left")

                        # GROUP BY COMMUNE AND CODE ROME AND COUNT JOB OFFERS
                        df = df.groupby(["commune", "fap87"]).size().reset_index(name="job_offers_fap87")

                        # CREATE MONTH COLUMN
                        date_obj = datetime.datetime.strptime(month_folder, "%Y-%m")
                        month_name = date_obj.strftime("%B")
                        df["month"] = month_name

                        # Append processed DataFrame to list
                        jocas_list.append(df)

                print(f"  Finished processing month folder: {month_folder}")

        print(f"Finished processing source folder: {source_folder}")

# CONCATENATE ALL DATAFRAMES
final_df = pd.concat(jocas_list, ignore_index=True)

# GROUP AGAIN TO ENSURE AGGREGATION ACROSS ALL FILES
final_df = final_df.groupby(["commune", "fap87", "month"], as_index=False).agg({"job_offers_fap87": "sum"})

# EXPORT FINAL DATASET
final_df.to_csv(output_path, index=False)

print(f"Processing complete. Final dataset saved to: {output_path}")
````

````python
# Distribution of "job_offers_fap87"
plt.figure(figsize=(8, 5))
final_df["job_offers_fap87"].hist(bins=300, edgecolor="red")
plt.xlabel("Job Offers (FAP87)")
plt.ylabel("Frequency")
plt.title("Distribution of Job Offers (FAP87)")
plt.show()
````

````python
# Group by month and compute total number of vacancies
job_month = final_df.groupby(["month"], as_index=False).agg({"job_offers_fap87": "sum"})

# Define the correct chronological order for months
month_order = ["January", "February", "March", "April", "May", "June", "July", 
               "August", "September", "October", "November", "December"]

# Ensure chronological order of months
job_month["month"] = pd.Categorical(job_month["month"], categories=month_order, ordered=True)

# Sort and plot
job_month = job_month.sort_values("month")
plt.figure(figsize=(10, 6))
plt.plot(job_month["month"], job_month["job_offers_fap87"], marker="o", linestyle="-", color="royalblue")

# Formatting
plt.xlabel("Month"), plt.ylabel("Number of Job Offers")
plt.title("Total Job Offers per Month"), plt.xticks(rotation=45), plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
````

‚ùó This figure on job offers in major French communes (2020) shows similar patterns to the STATISTA report on total job vacancies in France for the same year. However, our values are higher, which can be explained by several factors:

- Data sources: The figure below includes only job offers from the Employment Service (France Travail), while ours incorporates additional online sources (JOCAS), leading to a broader count.
- Mapping ROME to FAP codes: Transitioning from ROME to FAP codes introduces some duplication of job offers, as a single ROME code can correspond to multiple FAP codes. We assume that job offers linked to the same ROME code can be matched with any corresponding FAP, which may inflate the total count.



````python
# Aggregate job offers at FAP22 level and compute share
job_fap22 = (final_df.groupby("fap87", as_index=False)["job_offers_fap87"].sum()
             .merge(rome_fap, on="fap87", how="left")
             .drop_duplicates("fap87")
             .groupby("famille_pro22", as_index=False)["job_offers_fap87"].sum())

job_fap22["share_of_offers"] = job_fap22["job_offers_fap87"] / job_fap22["job_offers_fap87"].sum() * 100
job_fap22 = job_fap22.sort_values("share_of_offers")

# Plot
plt.figure(figsize=(10, 11))
bars = plt.barh(job_fap22["famille_pro22"], job_fap22["share_of_offers"], color="royalblue")

# Labels and Titles
plt.xlabel("Share of Job Offers (%)", fontsize=14)
plt.ylabel("FAP22 Categories", fontsize=14)
plt.title("Share of Job Offers by FAP22 Category", fontsize=16)

# Add percentage labels next to each bar
for bar in bars:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
             f"{bar.get_width():.1f}%", va="center", fontsize=12, color="black")

plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.show()
````
